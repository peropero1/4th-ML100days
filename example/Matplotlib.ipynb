# matplotlib有2種不同的介面：
## 1. matlab form: plt.plot()
## 2. 物件導向form: ax.plot()
    當圖表複雜時物件導向較好用，大部份情況下只是在plt.plot()和ax.plot()間切換

```python
# matlab form: subplot()

import numpy as np
%matplotlib inline

plt.figure()    # 建立一張figure

plt.subplot(2,1,1)  # 建立2個panel中的第一個，參數順序為(rows,columns, panel number)
plt.plot(x,np.sin(x),'-')

plt.subplot(2,1,2)  # 建立2個panel中的第2個
plt.plot(x,np.cos(x),'--')
```

```python
# object oriented form: ax.plot()
'''
matplotlib.pyplot.subplots(nrows=1, ncols=1, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)
    Create a figure and a set of subplots.

Returns:
    fig: Figure
    ax : axes.Axes object or array of Axes objects.
'''

import numpy as np
%matplotlib inline

fig,ax=plt.subplots(2)
    # 產生一個包含2 rows 1 cols(by default)的圖
    # ax 是2個軸物件的陣列

# 在適當的物件上呼叫plot方法
ax[0].plot(x,np.sin(x),'-')
ax[1].plot(x,np.cos(x),'--')
    # 結果同上
```

# matplotlib 基本概念：
|Item|說明|
|:-----|:-----|

|axes|plt.Axes的instance, 一個有邊界和刻度及label的方框 |

# 設定線條的顏色

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,1000)

plt.plot(x,np.sin(x-0),color='blue')  # 用名字
plt.plot(x,np.sin(x-1),color='g')     # 用顏色編碼
plt.plot(x,np.sin(x-2),color='0.75')  # 用灰階,0~1
plt.plot(x,np.sin(x-3),color='#FFDD44')  # 用16進位編碼
plt.plot(x,np.sin(x-4),color=(1.0, 0.2, 0.3))  # 用RGB tuple, 0~1
plt.plot(x,np.sin(x-5),color='chartreuse')  # 用html中支援的名稱
```

# 指定線條樣式
    linestyle or ls: {'-', '--', '-.', ':', '', (offset, on-off-seq), ...}

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,1000)

plt.plot(x,x+0,linestyle='-')
plt.plot(x,x+1,linestyle='--')
plt.plot(x,x+2,linestyle='-.')
plt.plot(x,x+3,linestyle=':')
    # 沒指定color的話會由一組預設的顏色循環使用
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,1000)

plt.plot(x,x+0,linestyle='-',color='green')
plt.plot(x,x+1,linestyle='--',color='yellow')
plt.plot(x,x+2,linestyle='-.',color='red')
plt.plot(x,x+3,linestyle=':',color='black')
```

# 調整Axes範圍
## 1.
    plt.xlim()：xlim(left, right)     # set the xlim to left, right
    plt.ylim()：ylim(left, right)   # set the ylim to left, right
    
## 2.
    plt.axis([xmin,xmax,ymin,ymax])

```python
# plt.xlim() & plt.ylim()

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-10,10,1000)
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

plt.subplot(2,1,2)
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
    #設定下圖的axes的x,y範圍，可和預設比較
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
# plt.axis([xmin,xmax,ymin,ymax])

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-10,10,1000)
plt.axis([-1,11,-1.5,1.5])

plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
# plt.axis('equal'): 讓x與y軸的比例相同
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,1000)
plt.axis('equal')

plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

# 為figure加上label
    plt.title(): Set a title for the axes
    plt.xlabel()
    plt.ylabel()

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x))
plt.title('A Sin Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
# plt.legend(): 圖標
#有時候legend會擋到圖檔，所以可以利用loc來調整位置，0=讓matplotlib決定最好的地方、1=右上、2=左上、3=左下、4=右下
# ex: plt.legend(loc=1)

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,1000)

plt.plot(x,np.sin(x),'-g',label='sin(x)')
plt.plot(x,np.cos(x),':b',label='cos(x)')

plt.legend()
```

# 散佈圖
    plt.plot(): 大資料集下效能較高, 因為每個點都是複製出來的
    plt.scatter()：有能力針對每個點進行繪製，可畫較精細但效能較差

```python
# plt.plot()

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,30)
y=np.sin(x)

plt.plot(x,y,'o',color='black')
```

```python
# 合併線條與點

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,30)
y=np.sin(x)

plt.plot(x,y,'-o',color='black')
    # 把樣式合併為'-o'
```

```python
# 各種不同形狀的點

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
rng=np.random.RandomState(0)

for marker in ['o','.',',','x','+','v','^','<','>','s','d','p']:
    plt.plot(rng.rand(5),rng.rand(5),marker,label="marker={}".format(marker))
plt.legend()
plt.xlim(0,2)
```

```python
# plt.plot()下針對點的其它參數

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,30)
y=np.sin(x)

plt.plot(x,y,'-p',color='grey',
        markersize=15,
        linewidth=4,
        markerfacecolor='yellow',
        markeredgecolor='green',
        markeredgewidth=2)
plt.xlim(-1,10)
```
